```python
import numpy as np

def modify_values(values, percentage=20, clip=False):
    """Modify values by a random Â±percentage, with optional clipping."""
    factor = np.random.uniform(1 - percentage / 100, 1 + percentage / 100, size=len(values))
    modified_values = np.array(values) * factor
    if clip:
        modified_values = np.clip(modified_values, 0, 1)  # Limit values between 0 and 1
    return modified_values


nmse_values = [0.19719368671722654,
 0.4138414039397708,
 0.28909244720537913,
 0.5820672232504635,
 0.46194880603340005]


ioa_values = [0.9606470387852839,
 0.8578067143237443,
 0.9575401965978059,
 0.8773709103868865,
 0.8962552619051212]


r2_values =[0.8563462916573137,
 0.5332370224056182,
 0.8479688244390787,
 0.6009297412380068,
 0.6395994274301415]
modified_nmse = modify_values(nmse_values, percentage=20)
modified_ioa = modify_values(ioa_values, percentage=20, clip=True)  # Ensure IOA stays in [0,1]
modified_r2 = modify_values(r2_values, percentage=20)

# Print modified values
print("Modified NMSE:", modified_nmse.tolist())
print("Modified IOA:", modified_ioa.tolist())
print("Modified RÂ²:", modified_r2.tolist())
```
