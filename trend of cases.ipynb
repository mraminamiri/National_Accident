```python
data = pd.read_csv('final_with_covid_AS.csv')
```

```python
filtered_data = data[data['CarDriver'] == 1]
filtered_data.info()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


filtered_data['Year'] = (
    filtered_data['Y2011'] * 2011 +
    filtered_data['Y2012'] * 2012 +
    filtered_data['Y2013'] * 2013 +
    filtered_data['Y2014'] * 2014 +
    filtered_data['Y2015'] * 2015 +
    filtered_data['Y2016'] * 2016 +
    filtered_data['Y2017'] * 2017 +
    filtered_data['Y2018'] * 2018 +
    filtered_data['Y2019'] * 2019 +
    filtered_data['Y2020'] * 2020 +
    filtered_data['Y2021'] * 2021
)

yearly_sum = filtered_data.groupby('Year')['CountCases'].sum().reset_index()

plt.figure(figsize=(10, 5))
plt.plot(yearly_sum['Year'], yearly_sum['CountCases'], marker='o')
plt.title('Trend of Case Counts CarDriver from 2011 to 2021')
plt.xlabel('Year')
plt.ylabel('Count of Cases CarDriver')
plt.grid(True)
plt.show()
```
